### Número total funcionarios en #3de3

```python
funcionarios = []
pagina = 1
while True:
    a = requests.get("https://3de3.mx/declarantes.json?page="+str(pagina))
    b = pd.DataFrame(a.json())
    funcionarios.append(b)
    pagina += 1
    if b.size == 0:
        break
```

```python
funcionarios = pd.concat(funcionarios).reset_index()
funcionarios
```

```python
funcionarios.count()
```

### Participación en #3de3 por género

```python
participacion = []
paginas = list(range(1,32))
genero = ["F","M"]
for p in paginas:
    for g in genero:
        r = requests.get("https://3de3.mx/declarantes.json?page="+str(p)+"&by_gender="+g)
        df_genero = pd.DataFrame(r.json())
        df_genero["gender"] = g
        participacion.append(df_genero)
        
participacion = pd.concat(participacion).reset_index()
participacion.count()
```

```python
participacion.groupby(["gender"]).count()
```

### Numero de funcionarios por partido

```python
funcionarios = []
paginas = list(range(1,32)) #this line is wrong, yet it worked.
partidos = ['PRI', 'PAN', 'PT', 'PVEM', 'PRD', 'MC', 'Nueva Alianza', 'Morena', 'PH', 'PES', 'Otro', 'Independiente']
for pagina in paginas:
    for partido in partidos:
        re = requests.get("https://3de3.mx/declarantes.json?page="+str(pagina)+"&by_political_party="+partido)
        df_funcionarios = pd.DataFrame(re.json())
        df_funcionarios["party"] = partido
        funcionarios.append(df_funcionarios)
        
funcionarios = pd.concat(funcionarios).reset_index()
```

```python
funcionarios = pd.concat(funcionarios).reset_index()
```

```python
funcionarios.groupby(["party"]).count()
#Tomar el index como referencia del número total.
```
